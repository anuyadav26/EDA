```python
import re

# Q14) Correlation between car price, engine size, and annual income

# First, we need to extract numeric engine size from the 'Engine' column if possible.
# Let's try to extract engine displacement (e.g., '2.0L', '2000cc', etc.) if present, otherwise this will be skipped.


def extract_engine_size(engine_str):
    # Try to extract numbers from engine string (e.g., '2.0L', '2000cc')
    match = re.search(r'(\d+(\.\d+)?)\s*(L|l|CC|cc|c|C)?', engine_str)
    if match:
        value = float(match.group(1))
        # Convert to liters if cc is detected
        if match.group(3) and 'cc' in match.group(3).lower():
            return value / 1000  # convert cc to liters
        return value
    return None

# Apply extraction to the Engine column
df['Engine_Size_L'] = df['Engine'].apply(extract_engine_size)

# Prepare the DataFrame for correlation analysis
corr_df = df[['Price ($)', 'Annual Income', 'Engine_Size_L']].copy()

# Drop rows with missing engine size if any
corr_df = corr_df.dropna()

# Compute correlation matrix
corr_matrix = corr_df.corr()
print("Correlation matrix between car price, engine size (L), and annual income:")
print(corr_matrix)

# Visualize the correlation matrix
plt.figure(figsize=(5, 3))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.tight_layout()
plt.show()

# Pairplot to visualize interactions
sns.pairplot(corr_df)
plt.suptitle('Pairwise Relationships: Price, Engine Size, Annual Income', y=1.02)
plt.show()
```
